```python
# definir visualização
mapa.choropleth(
    data_frame = df,
    geo_data = geo_json_data,
    name = 'Average Preparation Time per state',
    data = df,
    columns = ['UF', 'AVG'],
    key_on = 'feature.id',
    fill_color = 'YlOrRd',
    fill_opacity = 0.7,
    line_color = 'white',
    line_opacity = 1.0,
    show = True,
    range_color = [0, 5],
    nan_fill_color = 'gray',
    legend_name = 'Average Preparation Time per state'
)
```
